```python
#1. Unordered Collection
my_set = {1, 2, 3}
print(my_set)  # Output could be {1, 2, 3} or {3, 2, 1} (order not guaranteed)

#2. Unique Elements
# A set automatically removes any duplicate elements. Each element in a set must be unique.
# Example:
my_set = {1, 2, 2, 3}
print(my_set)  # Output: {1, 2, 3} (duplicate '2' is removed)

#3. Mutable
# Although the elements of a set are immutable (like integers, strings, etc.), you can add or remove elements from the set after its creation.
# Example:
my_set = {1, 2, 3}
my_set.add(4)  # Adds an element to the set
print(my_set)  # Output: {1, 2, 3, 4}

# Unindexed
# Sets do not support indexing because they are unordered. You cannot access elements by position as you would in lists or tuples.
# Example:
my_set = {1, 2, 3}
# my_set[0]  # Raises a TypeError because sets are unindexed

# Efficient Membership Testing
# Sets are highly optimized for membership testing, meaning checking if an element is present in the set is much faster compared to lists.
# Example:

my_set = {1, 2, 3}
print(2 in my_set)  # Output: True (quick membership test)

# No Duplicate Elements
# Sets automatically eliminate duplicate entries when created, making them useful for deduplicating lists or other sequences.
# Example:
my_list = [1, 2, 2, 3, 4, 4]
unique_set = set(my_list)
print(unique_set)  # Output: {1, 2, 3, 4}

# Set Operations
# Sets support common mathematical set operations like union, intersection, difference, and symmetric difference.
# Example:

set1 = {1, 2, 3}
set2 = {3, 4, 5}

# Union (all elements from both sets)
print(set1 | set2)  # Output: {1, 2, 3, 4, 5}

# Intersection (common elements in both sets)
print(set1 & set2)  # Output: {3}

# Difference (elements in set1 but not in set2)
print(set1 - set2)  # Output: {1, 2}

# Symmetric Difference (elements in either set, but not both)
print(set1 ^ set2)  # Output: {1, 2, 4, 5}

# Set Operations
# Sets support common mathematical set operations like union, intersection, difference, and symmetric difference.

# Mutable Set Methods
# add(): Adds an element to the set.
# remove(): Removes a specified element from the set (raises an error if the element is not present).
# discard(): Removes a specified element (does not raise an error if the element is not present).
# pop(): Removes and returns an arbitrary element from the set.
# clear(): Removes all elements from the set.

# Frozen Sets
# Python provides a frozen set, which is an immutable version of a set. Once created, you cannot modify a frozen set (no adding or removing elements). It can still be used for membership testing and set operations.

# examples of there use cases
# 1. Removing Duplicates from a List
# Sets are an easy way to remove duplicates from a list.

my_list = [1, 2, 2, 3, 4, 4]
unique_set = set(my_list)
print(unique_set)  # Output: {1, 2, 3, 4}

#  Membership Testing
# Sets are ideal for checking membership quickly.
my_set = {1, 2, 3, 4, 5}
print(3 in my_set)  # Output: True
print(6 in my_set)  # Output: False

# . Mathematical Set Operations
# Sets are commonly used to perform union, intersection, and difference operations.
evens = {2, 4, 6, 8}
odds = {1, 3, 5, 7}

all_numbers = evens | odds  # Union
print(all_numbers)  # Output: {1, 2, 3, 4, 5, 6, 7, 8}

# . Finding Common Elements
# To find the common elements between two collections, sets are highly efficient.
set1 = {1, 2, 3, 4}
set2 = {3, 4, 5, 6}
common_elements = set1 & set2  # Intersection
print(common_elements)  # Output: {3, 4}
```
